```python
import pandas as pd
import os

#
data = {
    "Height_in": [65.8, 71.5, 69.4, 68.2, 67.8, 68.7, 69.8, 70.1, 67.9, 66.8],
    "Weight_lb": [112, 136, 153, 142, 144, 123, 141, 136, 112, 120],
    "Age_yr": [30, 19, 45, 22, 29, 50, 51, 23, 17, 39],
    "Grip_kg": [30, 31, 29, 28, 24, 26, 22, 20, 19, 31],
    "Frailty": ["N", "N", "N", "Y", "Y", "N", "Y", "Y", "N", "N"]
}
df_raw = pd.DataFrame(data)
csv_file = "frailty_data.csv"
df_raw.to_csv(csv_file, index=False)
print(f"CSV file saved as '{csv_file}'")

#
df = pd.read_csv(csv_file)
print("\nData ingested successfully!")
print(df.head())


# a. Unit standardization
df["Height_m"] = df["Height_in"] * 0.0254
df["Weight_kg"] = df["Weight_lb"] * 0.45359237

# b. Feature engineering
df["BMI"] = round(df["Weight_kg"] / (df["Height_m"] ** 2), 2)

def age_group(age):
    if age < 30:
        return "<30"
    elif 30 <= age <= 45:
        return "30–45"
    elif 46 <= age <= 60:
        return "46–60"
    else:
        return ">60"

df["AgeGroup"] = df["Age_yr"].apply(age_group)

# c. Categorical → numeric encoding
df["Frailty_binary"] = df["Frailty"].map({"Y":1, "N":0}).astype("int8")
df = pd.get_dummies(df, columns=["AgeGroup"], prefix="AgeGroup")

print("\nProcessed DataFrame:")
print(df.head())

# --- 4. Stage 3: Analyze ---

os.makedirs("reports", exist_ok=True)

# I. Summary statistics
numeric_cols = ["Height_m", "Weight_kg", "Age_yr", "Grip_kg", "BMI"]
summary = df[numeric_cols].agg(["mean", "median", "std"]).round(2)
summary_md = summary.to_markdown()
with open("reports/findings.md", "w") as f:
    f.write("# Summary Statistics\n\n")
    f.write(summary_md)
    f.write("\n\n")

# II. Correlation Grip_kg ↔ Frailty_binary
correlation = df["Grip_kg"].corr(df["Frailty_binary"])
with open("reports/findings.md", "a") as f:
    f.write(f"## Grip Strength vs Frailty Correlation\n\n")
    f.write(f"Correlation between Grip_kg and Frailty_binary: {correlation:.2f}\n")

print("\nEDA report saved to 'reports/findings.md'")
print(f"Correlation between Grip_kg and Frailty_binary: {correlation:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="Set2")

from google.colab import files
uploaded = files.upload()


import io
file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_name]))


print("Columns:", df.columns)


df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")


df["math_score"] = df["math_score"].fillna(df["math_score"].mean())
df["reading_score"] = df["reading_score"].fillna(df["reading_score"].mean())
df["writing_score"] = df["writing_score"].fillna(df["writing_score"].mean())


df["overall_avg"] = df[["math_score","reading_score","writing_score"]].mean(axis=1)

print(df.head())
```

```python
plt.figure(figsize=(8,6), dpi=100)


sns.boxplot(x="gender", y="math_score", data=df)
sns.boxplot(x="gender", y="reading_score", data=df)

plt.title("Math vs Reading by Gender", fontsize=14)
plt.ylabel("Score")
plt.xlabel("Gender")

plt.savefig("V1_gender_boxplot.png", dpi=300, bbox_inches="tight")
plt.show()

print("""
Interpretation V1:
The boxplots show clear gender differences in performance across subjects.
Female students tend to score higher in reading compared to male students.
Conversely, male students generally achieve higher scores in math.
The median reading score for females is noticeably above that of males, while the reverse is true for math.
The spread of scores, shown by the interquartile range, appears similar between genders.
Overall, the results suggest that gender differences exist in academic strengths, with females excelling in reading and males in math.
""")
```

```python
plt.figure(figsize=(8,6), dpi=100)


sns.boxplot(x="test_preparation_course", y="math_score", data=df)

plt.title("Math Score by Test Preparation", fontsize=14)
plt.ylabel("Math Score")
plt.xlabel("Test Prep Course")

plt.savefig("V2_testprep_math.png", dpi=300, bbox_inches="tight")
plt.show()

print("""
Interpretation V2:
The chart illustrates the relationship between test preparation and math performance.
Students who completed the test preparation course achieved higher median math scores compared to those who did not.
The difference is not only in the median but also in the distribution, as scores are shifted upward overall.
This suggests that completing test preparation may provide a measurable benefit in math outcomes.
However, some overlap exists between the two groups, meaning test preparation is not the only factor influencing performance.
Still, the general pattern supports the effectiveness of preparation courses.
""")
```

```python
# C.V3 — Lunch type and average performance

plt.figure(figsize=(8,6), dpi=100)
sns.barplot(x="lunch", y="overall_avg", data=df, errorbar=None)
plt.title("Average Performance by Lunch Type", fontsize=14)
plt.ylabel("Overall Average Score")
plt.xlabel("Lunch Type")
plt.savefig("V3_lunch_avg.png", dpi=300, bbox_inches="tight")
plt.show()

print("""
Interpretation V3:
The bar chart compares average performance based on lunch type.
Students receiving a standard lunch score higher on overall averages than those with free or reduced lunch.
This difference is consistent across math, reading, and writing combined.
The results may reflect broader socioeconomic differences, as lunch type often serves as a proxy for family background.
While the gap is not extremely large, it is noticeable and significant enough to suggest that economic support can affect academic performance.
These findings highlight the link between nutrition, socioeconomic status, and educational outcomes.
""")
```

```python
plt.figure(figsize=(8,6), dpi=100)


corr = df[["math_score","reading_score","writing_score"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", vmin=0.5, vmax=1)
plt.title("Correlation Heatmap of Subjects", fontsize=14)
plt.savefig("V4_subject_corr.png", dpi=300, bbox_inches="tight")
plt.show()

print("""
Interpretation V4:
The heatmap shows strong positive correlations between math, reading, and writing scores.
Reading and writing have the highest correlation, indicating that students who perform well in reading typically also excel in writing.
Math also shows moderate to strong correlations with both reading and writing, though slightly weaker than the reading–writing relationship.
This pattern suggests that academic skills tend to move together rather than separately.
The results imply that interventions to improve one subject may have spillover effects on the others.
Overall, the data demonstrates that the three core subjects are closely interconnected.
""")
```
